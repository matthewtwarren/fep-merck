```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import AllChem
from rdkit.Chem import DataStructs
import mols2grid
import MDAnalysis as mda
```

```python
def extract_ligand_from_pdb(pdb_file, resname):

    u = mda.Universe(pdb_file)
    ligand = u.select_atoms("resname " + resname)
    ligand.write(pdb_file.replace(".pdb", "_ligand.pdb"))
    
```

```python
def pdb_to_smiles(pdb_file):
    
    mol = Chem.MolFromPDBFile(pdb_file)
    smiles = Chem.MolToSmiles(mol)
    return smiles
```

```python
def sdf_to_mols(sdf_filepath):

    sdf_supplier = Chem.SDMolSupplier(sdf_filepath, removeHs=False)
    ligand_dict = {}
    
    for mol in sdf_supplier:
        if mol is not None:
            name = mol.GetProp("_Name") if mol.HasProp("_Name") else "Unnamed"
            smiles = Chem.MolToSmiles(mol)
            ligand_dict[name] = mol
            
    return ligand_dict
```

```python
def draw_mol(smiles):
        
    mol = Chem.MolFromSmiles(smiles)
    return Draw.MolToImage(mol)
```

```python
# CDK8
extract_ligand_from_pdb("cdk8/5hnb.pdb", "62M")
crystal_smiles = pdb_to_smiles("cdk8/5hnb_ligand.pdb")
draw_mol(crystal_smiles)
```
